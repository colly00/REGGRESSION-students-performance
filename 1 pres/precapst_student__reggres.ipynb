```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Load dataset
df = pd.read_csv(r'C:\Users\TFC\Downloads\student_habits_performance.csv')
df
```

CLEANING DATA

```python
# Drop ID column
df.drop('student_id', axis =1, inplace=True)
df.drop('age', axis =1, inplace=True)
df.drop('diet_quality', axis =1, inplace=True)
df.drop('exercise_frequency', axis =1, inplace=True)
df.drop('parental_education_level', axis =1, inplace=True)
df.drop('internet_quality', axis =1, inplace=True)
df.drop('mental_health_rating', axis =1, inplace=True)
```

```python
df.drop('binned', axis =1, inplace=True)
```

```python
# Encode all categorical (object) columns into numeric
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoders = {}
for col in df.select_dtypes(include="object").columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df.drop('exam_score', axis=1)
y = df['exam_score']

# Split data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30 )
```

```python
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize and train the model
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred= model.predict(X_test_scaled)
# Predict on test set
```

```python
# Evaluate the model
rmse = mean_squared_error(y_test, y_pred, squared=False)
lr_rmse_percent=(rmse/np.mean(y_test))*100
r2 = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error (RMSE):{lr_rmse_percent:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
from sklearn.linear_model import LassoCV
lasso = LassoCV(cv=5, random_state=42)
lasso.fit(X_train_scaled, y_train)

# Get feature importance
importance = pd.Series(lasso.coef_, index=X_train.columns)
importance_sorted = importance.sort_values(ascending=False)

print("\nðŸ§  Feature Importance from LassoCV:")
print(importance_sorted)

# Optionally: visualize
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
importance_sorted.plot(kind='barh')
plt.title("Feature Importance (Lasso)")
plt.xlabel("Coefficient Weight")
plt.tight_layout()
plt.show()
```
